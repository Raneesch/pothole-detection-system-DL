```python
import threading
from ultralytics import YOLO
import cv2
import numpy as np
import cvzone
import pyttsx3
```

```python
# Load model

model = YOLO('best.pt')
class_name =  model.names
cap = cv2.VideoCapture('video3.mp4')

```

```python


# Initialize the text-to-speech engine
engine = pyttsx3.init()

# Function to play voice alert in a separate thread
def play_voice_alert(message):
    def speak():
        engine.say(message)
        engine.runAndWait()
    alert_thread = threading.Thread(target=speak)
    alert_thread.start()

count = 0
while True:
    ret, img = cap.read()
    if not ret:
        break
    count += 1
    if count % 3 != 0:
        continue

    img = cv2.resize(img, (1020, 500))
    h, w, _ = img.shape
    results = model.predict(img)

    pothole_detected = False  # Flag to track pothole detection

    for r in results:
        boxes = r.boxes  # boxes object for bbox output
        masks = r.masks  # masks object for segment masks output

    if masks is not None:
        masks = masks.data.cpu()
        for seg, box in zip(masks.data.cpu().numpy(), boxes):
            seg = cv2.resize(seg, (w, h))
            contours, _ = cv2.findContours((seg).astype(np.uint8), cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

            for contour in contours:
                d = int(box.cls)
                c = class_name[d]
                x, y, x1, y1 = cv2.boundingRect(contour)
                cv2.polylines(img, [contour], True, color=(0, 0, 255), thickness=2)
                # cv2.rectangle(img,(x,y),(x1+x,y1+y),color=(255,0,0),thickness=2)
                cv2.putText(img, c, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0))

                # Check if the detected class is a pothole
                if c.lower() == "pothole":
                    pothole_detected = True

    # Voice alert for pothole detection (non-blocking)
    if pothole_detected:
        play_voice_alert("Pothole detected!")

    # Display the count on the screen
    # cv2.putText(img, f"Frame Count: {count}", (10, 30), cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2)

    cv2.imshow('img', img)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```
